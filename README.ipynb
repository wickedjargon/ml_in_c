```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('spreadsheet.csv')

# Plot the guess and cost over iterations
fig, ax1 = plt.subplots()

ax1.set_xlabel('Iteration')
ax1.set_ylabel('Guess', color='tab:blue')
ax1.plot(data['Iteration'], data['Guess'], color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()  
ax2.set_ylabel('Cost', color='tab:red')  
ax2.plot(data['Iteration'], data['Cost'], color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.tight_layout()  
plt.title('Gradient Descent Optimization')
plt.show()
```
